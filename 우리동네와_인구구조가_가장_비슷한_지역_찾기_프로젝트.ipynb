# 3. 다른 지역과의 비교를 위해 비율(해당연령의 인구수/각지역의 전체인구수)로 표현


```python
import csv
f=open('age.csv')
data=csv.reader(f)

result=[]
result2=[]
for row in data:
  if '신도림' in row[0]:
    for i in range(3,len(row)):
      result.append(int(row[i].replace(',',''))/int(row[2].replace(',','')))   # Remove commas before converting to integer
  if '봉선1동' in row[0]:
    for i in range(3,len(row)):
      result2.append(int(row[i].replace(',',''))/int(row[2].replace(',','')))

import matplotlib.pyplot as plt
plt.plot(result)
plt.plot(result2)
plt.show()
```

# 4.  s= 기준 지역의 해당 연령값 - 각지역의 해당연령값캡처.JPG

```python
import csv
f = open('age.csv')

data = csv.reader(f)
next(data)
next(data)
data = list(data)
pivot = []
for row in data :
    if '신도림' in row[0] :
        for i in range(3,len(row)) :
            pivot.append(int(row[i].replace(',',''))/int(row[2].replace(',','')))
mn = 10000000

for row in data :
    # Check if the population is zero before dividing
    if int(row[2].replace(',', '')) != 0:
        s = 0
        for i in range(3, len(row)) :
            tmp = int(row[i].replace(',','')) - pivot[i-3]
            s = s + tmp
        if s < mn :
            result = []
            for i in range(3, len(row)) :
                # Perform the division if population is not zero
                result.append(int(row[i].replace(',',''))/int(row[2].replace(',','')))


import matplotlib.pyplot as plt
plt.plot(pivot)
plt.plot(result)
plt.show()
```

# 5. 캡처.JPG

```python
import csv
f = open('age.csv')

data = csv.reader(f)
next(data)
next(data)
data = list(data)
pivot = []
for row in data :
    if '신도림' in row[0] :
        for i in range(3,len(row)) :
            pivot.append(int(row[i].replace(',',''))/int(row[2].replace(',','')))
mn = 1

for row in data :
    s = 0
    # Check if the population is zero before performing the division
    if int(row[2].replace(',', '')) != 0:
        for i in range(3, len(row)) :
            row[i] = int(row[i].replace(',',''))/int(row[2].replace(',',''))
            tmp = (row[i] - pivot[i-3]) ** 2
            s = s + tmp
    if s < mn :
        result = []
        for i in range(3, len(row)) :
            result.append(row[i])
        mn = s

import matplotlib.pyplot as plt
plt.plot(pivot)
plt.plot(result)
plt.show()
```

# 8. 범례 추가하기

```python
import csv
f = open('age.csv')

data = csv.reader(f)
next(data)
next(data)
data = list(data)
pivot = []

name = input('궁금한 지역 이름을 입력해주세요 : ')

for row in data :
    if name in row[0] :
        for i in range(3,len(row)) :
            # Remove commas before converting to integers
            pivot.append(int(row[i].replace(',',''))/int(row[2].replace(',','')))
mn = 1

for row in data :
    s = 0
    # Check if population is zero before calculating ratios
    if int(row[2].replace(',', '')) != 0:
        for i in range(3, len(row)) :
            # Remove commas before converting to integers
            row[i] = int(row[i].replace(',',''))/int(row[2].replace(',',''))
            tmp = (row[i] - pivot[i-3]) ** 2
            s = s + tmp
    if s < mn and (name not in row[0]) :
        result = []
        for i in range(3, len(row)) :
            result.append(row[i])
        mn = s
        result_name = row[0]

import matplotlib.pyplot as plt
plt.plot(pivot, label = name)
plt.plot(result, label = result_name)
plt.legend()
plt.show()
```
