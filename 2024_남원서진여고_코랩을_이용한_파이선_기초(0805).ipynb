##range
range(x) : 0~x 미만   
range(x,y) : x이상 y미만

```python
import random
lot = range(46)   #0~45까지의 수
print(random.choice(lot))
```

```python
import random
lot = range(46)   #0~45까지의 수
print('숫자 6개는:',random.sample(lot, 6))
```

```python
import random
lot = range(1, 46)   #1~45까지의 수
print(random.choice(lot))
print('로또 예상 번호는:',random.sample(lot, 6))
```

```python
#중첩 if
import random
com=random.randint(1,10)
user=int(input('숫자를 입력하세요'))
if user>com:
  print('그 보다 큰 수 입니다.')
elif user<com:
  print('그 보다 작은 수 입니다.')
else:
  print('정답입니다.')
```

```python
#중첩 if
import random
com=random.randint(1,10)
while True:
  user=int(input('1~10까지 어떤 수가 나올까요?'))
  if com>user:
    print('그 보다 큰 수 입니다.')
  elif com<user:
    print('그 보다 작은 수 입니다.')
  else:
    print('정답입니다.')
    break
```

**세수를 입력받아 최대값과 최소값을 구하는 프로그램을 구현하세요.**

```python
a=input('수를 입력하세요')
b=input('수를 입력하세요')
c=input('수를 입력하세요')
a=int(a)
b=int(b)
c=int(c)
if a>=b and a>=c:
  print(a)
elif b>=a and b>=c:
  print(b)
else:
  print(c)
```

```python
a, b, c=input('세 수를 입력하세요').split()
a=int(a)
b=int(b)
c=int(c)
if a>=b and a>=c:
  print(a)
elif b>=a and b>=c:
  print(b)
else:
  print(c)
```

```python
a, b, c=map(int,input('세 수를 입력하세요').split())

if a>=b and a>=c:
  print(a)
elif b>=a and b>=c:
  print(b)
else:
  print(c)
```
